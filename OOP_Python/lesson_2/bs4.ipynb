### Получение текста из HTML.

Находим и вырезаем все теги из файла

```python
with open("rbk.html") as f:
    s = f.read()
```

```python
tags = re.compile("<[^<]*>")
s = tags.sub("", s)
print(s)
```

Вырезаем комментарии и скрипты

```python
comments = re.compile("<!--.*?-->", re.DOTALL)
scripts = re.compile("<script.*?</script>", re.DOTALL)
s = comments.sub("", s)
s = scripts.sub("", s)


print(s)
```

Добавим строки, которые почистят пробельные символы

```python
s = re.sub("[ ]+"," ", s)
s = re.sub("\s{2,}", "\n", s)
print(s)
```

### Библиотека BeautifulSoup.

```python
from bs4 import BeautifulSoup as BS
```

```python
with open("rbk.html") as f:
    s = f.read()
```

```python
soup = BS(s, "html.parser")
```

```python
soup.title.string #выводим то, что содержится в заголовке сайта
```

```python
soup.div["class"] #получаем атрибуты
```

```python
soup.find_all("div") #ищем все элементы, которые содержат тег <div>
```

```python
len(soup.find_all("div"))
```
